```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define base directories and file structure for each dataset
datasets = [
    {
        "base_dir": 'data/rmsd/1oga/',
        "raw_color": "gray",
        "avg_color": "black", # Black
    },
    {
        "base_dir": 'data/rmsd/7kgq/',
        "raw_color": "#DB8557",
        "avg_color": "#D1662E", # Brown
    },
    {
        "base_dir": 'data/rmsd/7rtr/',
        "raw_color": "#D85A86",
        "avg_color": "#CF3068", # Red
    },
    {
        "base_dir": 'data/rmsd/7rtr_LLY/',
        "raw_color": "#4C84E6",
        "avg_color": "#1F65E0", # Blue
    },
]

sub_dirs = ['res']
files_in_sub_dirs = ['proa', 'prob', 'proc', 'pocket']

# Define the window size for the running average
window_size = 10

# Define output directory for figures
output_dir = './'

# Define the default limits and ticks for the x and y axes
default_x_limits = [0, 500]
default_y_limits = [0, 6]
default_x_ticks = list(range(0, 501, 250))  # example ticks, modify as needed
default_y_ticks = list(range(0, 7, 2))  # example ticks, modify as needed

# Define font sizes
axis_fontsize = 18  # modify as needed
legend_fontsize = 10  # modify as needed

# Define individual limits and ticks for the x and y axes (if needed)
individual_settings = {
    'prob': {
        'limits': ([0, 500], [0, 12]),
        'ticks': ([0, 250, 500], [0, 4, 8, 12])  # example ticks, modify as needed
    },
    # add more as needed
}

n_rows = len(files_in_sub_dirs)
n_columns = len(sub_dirs)

for dataset in datasets:
    base_dir = dataset['base_dir']
    raw_color = dataset['raw_color']
    avg_color = dataset['avg_color']

    fig, axs = plt.subplots(n_rows, n_columns, figsize=(4*n_columns, 2*n_rows))
    
    if n_columns == 1:
        axs = np.expand_dims(axs, axis=1)
    
    for i, sub_dir in enumerate(sub_dirs):
        for j, file in enumerate(files_in_sub_dirs):
            file_path = base_dir + sub_dir + '-' + file
            label = file
            data = np.loadtxt(file_path)

            ax = axs[j, i]
            ax.plot(data[:, 1], alpha=0.5, color=raw_color)
            running_avg = pd.Series(data[:, 1]).rolling(window=window_size).mean()
            ax.plot(running_avg, label=label, color=avg_color)

            settings = individual_settings.get(label, {})
            x_limits, y_limits = settings.get('limits', (default_x_limits, default_y_limits))
            x_ticks, y_ticks = settings.get('ticks', (default_x_ticks, default_y_ticks))

            ax.set_xlim(x_limits)
            ax.set_ylim(y_limits)
            ax.set_xticks(x_ticks)
            ax.set_xticklabels(x_ticks, fontsize=axis_fontsize)
            ax.set_yticks(y_ticks)
            ax.set_yticklabels(y_ticks, fontsize=axis_fontsize)
            
            # If it is not the last subplot in the column, remove x axis labels
            if j < n_rows - 1:
                ax.set_xticklabels([])
            
#             ax.legend(loc='upper right', fontsize=legend_fontsize)

    plt.tight_layout()
    plt.savefig(f'{output_dir}figure_{base_dir.split("/")[-2]}.tif', format='tif')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
import os

# List of proteins to plot
proteins_to_include = ['GIL', 'LLL', 'YLQ', 'LLY']  # replace with your protein names

# Function to get file paths for a list of proteins
def get_file_paths(proteins, file_type):
    file_paths = []
    for protein in proteins:
        file_paths.extend(glob.glob(f"data/AtomExposed/{protein}/*_{file_type}.csv"))
    return file_paths

peptide_file_paths = get_file_paths(proteins_to_include, 'peptide')
hlahelix_file_paths = get_file_paths(proteins_to_include, 'hlahelix')

# specify columns to include
columns_to_include = ['O', 'N', 'CS']  # update this list as needed

# specify color for each column
colors = ['red', 'green', 'blue', 'purple']  # update this list as needed

# specify error cap size
capsize = 5  # update this as needed

# Function to calculate averages, standard deviations and plot
def plot_averages(file_paths, title, columns_to_include, colors, capsize):
    averages = []
    stds = []
    labels = []
    for path in file_paths:
        df = pd.read_csv(path)
        averages.append(df[columns_to_include].mean())
        stds.append(df[columns_to_include].std())
        # Extract protein name from file path
        protein_name = os.path.basename(os.path.dirname(path))
        labels.append(protein_name)

    # Convert list to DataFrame
    averages_df = pd.DataFrame(averages, index=labels)
    stds_df = pd.DataFrame(stds, index=labels)

    # Create figure and axis
    fig, ax = plt.subplots()

    # Plot stacked horizontal bar plot with error bars
    averages_df.plot(kind='barh', stacked=True, xerr=stds_df, ax=ax, color=colors, capsize=capsize)

    plt.xlabel('Average Value')
    plt.ylabel('Protein')
    plt.title(title)
    plt.legend(title='Elements', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# Call function for both types of files
plot_averages(peptide_file_paths, 'Averages of Each Element per Peptide File', columns_to_include, colors, capsize)
plot_averages(hlahelix_file_paths, 'Averages of Each Element per HLAHelix File', columns_to_include, colors, capsize)
```
